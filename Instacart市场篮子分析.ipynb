```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
```

```python
#读取四张表的数据
prior=pd.read_csv('./data/Instacart Market Basket Analysis/order_products__prior.csv')
```

```python
prior.shape
```

```python
products=pd.read_csv('./data/Instacart Market Basket Analysis/products.csv')
```

```python
orders=pd.read_csv('./data/Instacart Market Basket Analysis/orders.csv')
```

```python
aisles=pd.read_csv('./data/Instacart Market Basket Analysis/aisles.csv')
```

```python
#合并四张表到一张表当中  (用户 - 物品类别)
_mg=pd.merge(prior,products,on=['product_id','product_id'])
_mg=pd.merge(_mg,orders,on=['order_id','order_id'])
mt=pd.merge(_mg,aisles,on=['aisle_id','aisle_id'])
```

```python
mt.head(10)
```

```python
#交叉表（特殊的分组工具）
cross=pd.crosstab(mt['user_id'],mt['aisle'])
```

```python
cross.head(10)
```

```python
#进行主成分分析(PCA)
pca=PCA(n_components=0.9)
```

```python
data=pca.fit_transform(cross)
```

```python
data.shape
```

```python
# x=data[:1000]
# x.shape
```

```python
# 假设用户一共分为4个类别
km=KMeans(n_clusters=4)
```

```python
km.fit(data)
```

```python
predict=km.predict(data)
```

```python
predict
```

```python
# 显示聚类的结果
plt.figure(figsize=(16,8))
```

```python
# 建立四个颜色的列表
colored=['orange','green','blue','purple']
colr=[colored[i] for i in predict]
plt.scatter(data[:,1],data[:,20],color=colr)
```

```python
# 评判聚类效果，轮廓系数
silhouette_score(data,predict)
```
